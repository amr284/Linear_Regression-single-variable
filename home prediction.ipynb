# predict home price using machine learning

we will use machine learning technique called linear regression 

in the table i have the price of homes based on the area of the home 

i will build amachine learning model that can tell me the prices of the home whose area 

3300 square feet ...?

5000 square feet ....?

price(Dependant variable) = m*area(independant variable) + b

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# load the homeprices in the dataframe 
df = pd.read_csv("home prices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area (sqr ft)')
plt.ylabel('price (us$)')
plt.scatter(df.area,df.price , color='red' , marker='+')
```

```python
# create linear regretion object from linear_model  
reg = linear_model.LinearRegression()
# fit data: means training  the linear regression model using avilable data models  
reg.fit(df[['area']] , df.price)
# when it is excute the linear regression model is ready to predict the price 
```

(1) predict price of ahome with area = 3300 sqr  ft  

```python
# predict if the area is 3300
reg.predict([[3300]])
```

we wanted to predict was the price of a home whose area is 3300

 let see to internal details so when i train regression object using this available data

```python
# show slope
reg.coef_
```

```python
# the value of b 
reg.intercept_
```

```python
# y = m*x+b
y= 3300*reg.coef_ + reg.intercept_
y
```




(2)predict price of ahome with area = 5000 sqr ft  

```python
reg.predict([[5000]])
```

```python
# y=m*x+b
y=5000 *reg.coef_ + reg.intercept_
y
```

```python
# create dataframe 
d= pd.read_csv("areas.csv")
d.head(3)
```
