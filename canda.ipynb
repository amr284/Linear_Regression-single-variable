# Predict canada's per capita income in year 2021.

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the homeprices in the dataframe 
df = pd.read_csv("canada.csv")
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df.income,color='red',marker='+')
```

```python
year = df[['year']]
year
```

```python
income = df['income']
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(year,income)
```

```python
reg.predict([[2020]])
```

```python
predicted_income = reg.predict(year)
predicted_income
```

```python
predicted_df = pd.DataFrame(year)
predicted_df.head(3)
```

```python
predicted_df['predicted income'] = predicted_income
predicted_df.head(3)
```

```python
%matplotlib inline
plt.xlabel('year', fontsize=20)
plt.ylabel('per capita income (US$)', fontsize=20)
plt.scatter(df.year,df.income ,color='red',marker='+')
plt.plot(predicted_df['year'],predicted_df['predicted income'],color='blue')
```
